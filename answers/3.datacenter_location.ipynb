# OpenDC Demo 3
### Data Center Location

In the [experiment 1](1.first_experiment.ipynb), we learned how to run OpenDC experiments and how to analyze and visualize the results to learn about the behavior of a data center. In this demo, we are using OpenDC to determine the impact of horizontal scaling. Horizontal scaling refers to reducing (or increasing) the number of machines available in a data center to fit a workload better.

In this demo, we will examine how the location of a data center affects its sustainability. We will run the same workload on the same hardware, but in four different locations. 

# Topology

To compare data centers at different locations, we need to create more topology files, one for each location. 

#### Exercise 1:

# Experiment

Experiment files can be used to run multiple simulations, using a single file. 

In this demo we want to run a seperate simulation for each of the four data centers.

#### Exercise 2: 
Update the experiment file of this demo (located [here](experiments/2.datacenter_location/location_experiment.json)) to execute a simulation for each of the data centers. 

```python
import subprocess

pathToScenario = "experiments/3.datacenter_location/location_experiment.json"
subprocess.run(["OpenDCExperimentRunner/bin/OpenDCExperimentRunner", "--experiment-path", pathToScenario])
```

## Output

#### Exercise 3: 
Load the results into Pandas DataFrames

```python
import pandas as pd 
import matplotlib.pyplot as plt


df_host_at = pd.read_parquet("output/3.datacenter_location/raw-output/0/seed=0/host.parquet")
df_power_at = pd.read_parquet("output/3.datacenter_location/raw-output/0/seed=0/powerSource.parquet")
df_service_at = pd.read_parquet("output/3.datacenter_location/raw-output/0/seed=0/service.parquet")
df_task_at = pd.read_parquet("output/3.datacenter_location/raw-output/0/seed=0/task.parquet")

df_host_au = pd.read_parquet("output/3.datacenter_location/raw-output/1/seed=0/host.parquet")
df_power_au = pd.read_parquet("output/3.datacenter_location/raw-output/1/seed=0/powerSource.parquet")
df_service_au = pd.read_parquet("output/3.datacenter_location/raw-output/1/seed=0/service.parquet")
df_task_au = pd.read_parquet("output/3.datacenter_location/raw-output/1/seed=0/task.parquet")

df_host_be = pd.read_parquet("output/3.datacenter_location/raw-output/2/seed=0/host.parquet")
df_power_be = pd.read_parquet("output/3.datacenter_location/raw-output/2/seed=0/powerSource.parquet")
df_service_be = pd.read_parquet("output/3.datacenter_location/raw-output/2/seed=0/service.parquet")
df_task_be = pd.read_parquet("output/3.datacenter_location/raw-output/2/seed=0/task.parquet")

df_host_nl = pd.read_parquet("output/3.datacenter_location/raw-output/3/seed=0/host.parquet")
df_power_nl = pd.read_parquet("output/3.datacenter_location/raw-output/3/seed=0/powerSource.parquet")
df_service_nl = pd.read_parquet("output/3.datacenter_location/raw-output/3/seed=0/service.parquet")
df_task_nl = pd.read_parquet("output/3.datacenter_location/raw-output/3/seed=0/task.parquet")
```

## Visualization

#### Exercise 3: 
Plot the energy usage of the data centers at different locations

```python
fig, ax = plt.subplots(figsize=(15,5))

ax.plot(df_power_at.timestamp/1000/60/60/24, df_power_at.energy_usage / 1000, label="AT")
ax.plot(df_power_au.timestamp/1000/60/60/24, df_power_au.energy_usage / 1000, label="AU")
ax.plot(df_power_be.timestamp/1000/60/60/24, df_power_be.energy_usage / 1000, label="BE")
ax.plot(df_power_nl.timestamp/1000/60/60/24, df_power_nl.energy_usage / 1000, label="NL")

plt.title("Energy Usage on different locations")
ax.legend()

plt.xlabel("Time (days)")
plt.ylabel("Energy (kWh)")

plt.show()
```

#### Exercise 4: 
Plot the energy usage of the data centers at different locations

```python
fig, ax = plt.subplots(figsize=(15,5))

ax.plot(df_power_at.timestamp/1000/60/60/24, df_power_at.carbon_emission / 1000, label="AT")
ax.plot(df_power_au.timestamp/1000/60/60/24, df_power_au.carbon_emission / 1000, label="AU")
ax.plot(df_power_be.timestamp/1000/60/60/24, df_power_be.carbon_emission / 1000, label="BE")
ax.plot(df_power_nl.timestamp/1000/60/60/24, df_power_nl.carbon_emission / 1000, label="NL")

plt.title("Carbon Emission on different locations")
ax.legend()

plt.xlabel("Time (days)")
plt.ylabel("Carbon (kg)")

plt.show()
```

#### Exercise 5: 
Compare the total carbon emissions of the different data centers.

```python
carbon_AT = df_power_at.carbon_emission.sum() / 1000
carbon_AU = df_power_au.carbon_emission.sum() / 1000
carbon_BE = df_power_be.carbon_emission.sum() / 1000
carbon_NL = df_power_nl.carbon_emission.sum() / 1000

print(f"Total carbon emission in AT: {carbon_AT:.2f} kg")
print(f"Total carbon emission in AU: {carbon_AU:.2f} kg")
print(f"Total carbon emission in BE: {carbon_BE:.2f} kg")
print(f"Total carbon emission in NL: {carbon_NL:.2f} kg")
```
